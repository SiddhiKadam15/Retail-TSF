# Bivariate Analysis

```python
plt.figure(figsize= (8,6))
sns.histplot(binwidth=0.5, x="Segment", hue="Ship Mode", data=df, stat="count", multiple="stack")
plt.xticks(rotation=0)
plt.show()
```

#### The above graph shows that the maximum count is for Ship mode 'Standard Class' and minimum count is for Ship mode 'Same day'.

```python
plt.figure(figsize= (12,6))
sns.histplot(binwidth=0.5, x="State", hue="Category", data=df, stat="count", multiple="stack")
plt.xticks(rotation=90)
plt.show()
```

#### The above graph shows that the categories-Office supplies, Furniture and Technology are highest in the State California.

```python
plt.figure(figsize= (15,10))
sns.histplot(binwidth=0.5, x="Category", hue="Sub-Category", data=df, stat="count", multiple="stack")
plt.xticks(rotation=0)
plt.show()
```

#### The above graph shows that the Category 'Office supplies' has multiple Sub-categories.

```python
df.groupby('Category')['Discount'].sum().plot.bar()
```

#### The above graph shows that the Discount for Category 'Office supplies' is maximum and 'Technology' is minimum.

```python
df.groupby('Category')['Quantity'].sum().plot.bar()
```

#### The above graph shows that the Quantity sold for Category 'Office supplies' is maximum and 'Technology' is minimum.

```python
df.groupby('Category')['Profit'].sum().plot.bar()
```

#### The above graph shows that the Profit earned for Category 'Technology' is maximum and 'Furniture' is minimum.

```python
df.groupby('Sub-Category')['Profit'].sum().plot.bar()
```

#### The above graph shows that the Profit earned is negative ie Loss is obtained for Sub-Category 'Bookcases', 'Supplies' and 'Tables' and hence they need to be focused upon.

```python
df.groupby('Sub-Category')['Discount'].sum().plot.bar()
```

#### The above graph shows that the maximum discount is offered for Sub-Category 'Blinders' and minimum for 'Copiers'.

```python
df.groupby('Sub-Category')['Quantity'].sum().plot.bar()
```

#### The above graph shows that the maximum quantity sold is for Sub-Category 'Blinders' and minimum for 'Copiers'.

```python
plt.figure(figsize= (12,10))
df.groupby('State')['Sales'].mean().plot.bar()
plt.show()
```

#### The above graph shows that the maximum sales is done by State 'Wyoming' and minimum by 'South Dakota'.

```python
df.groupby('Region')['Sales'].mean().plot.bar()
plt.show()
```

#### The above graph shows that the maximum sales is done by Region 'South' and minimum sales is done by Region 'Central'.

```python
sns.countplot(data= df, x='Category', hue ='Segment')
```

#### The above graph shows that the maximum sales is done by the Category 'Office supplies' with the highest count being of Segment 'Consumer'.

```python
sns.countplot(data= df, x='Category', hue ='Ship Mode')
```

#### The above graph shows that the maximum sales is done by the Category 'Office supplies' with the highest count being of Ship mode 'Standard Class'.

# Multivariate Analysis

```python
df.groupby('Category')['Sales','Quantity','Discount','Profit'].sum().plot.bar()
```

#### The above graph shows that the sales is highest for Category 'Technology'.

```python
plt.figure(figsize= (20,15))
df.groupby('Sub-Category')['Sales','Quantity','Discount','Profit'].sum().plot.bar()
plt.show()
```

#### The above graph shows that the sales is highest for Sub-Category 'Phones'.

## Heatmap

```python
sns.heatmap(df[['Sales', 'Quantity', 'Discount', 'Profit']].corr(), annot= True, cmap = "Blues")
```

Loss is obtained for Sub-Category 'Bookcases', 'Supplies' and 'Tables' and hence they need to be focused upon.

The maximum sales is done by State 'Wyoming'.

The maximum sales is done by the Category 'Office supplies' with the highest count being of Segment 'Consumer'.
